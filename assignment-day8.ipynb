```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Dataset
data = {
    'Name': ['AKHIL', 'NIKHIL', 'POOJA', 'MEGHANA', 'HARISH'],
    'Age': [20, 22, 21, 23, 24]
}

df = pd.DataFrame(data)

# Line Plot
sns.lineplot(x=df.index, y='Age', data=df)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Student Age Trend')
plt.show()

# Scatter Plot
sns.scatterplot(x=df.index, y='Age', data=df)
plt.title('Age Scatter Plot')
plt.show()

# Box Plot
sns.boxplot(y='Age', data=df)
plt.title('Age Distribution Box Plot')
plt.show()

# Violin Plot
sns.violinplot(y='Age', data=df)
plt.title('Age Distribution Violin Plot')
plt.show()

# Swarm Plot
sns.swarmplot(x=df.index, y='Age', data=df)
plt.title('Age Swarm Plot')
plt.show()

# Bar Plot
sns.barplot(x='Name', y='Age', data=df)
plt.title('Student Age Comparison')
plt.show()

# Point Plot
sns.pointplot(x='Name', y='Age', data=df)
plt.title('Age Point Plot')
plt.show()

# Count Plot Dataset
data2 = {
    'Name': [
        'AKHIL', 'NIKHIL', 'AKHIL',
        'POOJA', 'MEGHANA', 'MEGHANA',
        'HARISH', 'NIKHIL', 'AKHIL'
    ]
}

df2 = pd.DataFrame(data2)

# Count Plot
sns.countplot(x='Name', data=df2)
plt.title("Student Name Frequency")
plt.show()

# Load Iris Dataset
from sklearn.datasets import load_iris

iris = load_iris()

df3 = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df3['Species'] = iris.target

df3['Species'] = df3['Species'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

# KDE Plot
sns.kdeplot(
    data=df3[df3['Species'] == 'Versicolor'],
    x='sepal length (cm)',
    fill=True,
    label='Versicolor'
)

plt.legend()
plt.title('KDE Plot of Sepal Length')
plt.show()

# Load Seaborn Iris Dataset
iris = sns.load_dataset('iris')

# Scatter Plot with Labels
sns.scatterplot(
    x='petal_length',
    y='petal_width',
    data=iris
)

plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

# Built-in Style
sns.set_style("darkgrid")

sns.boxplot(
    x='species',
    y='sepal_length',
    data=iris
)

plt.title('Sepal Length Distribution by Species')
plt.show()

# Built-in Color Palette
sns.set_palette("muted")

sns.violinplot(
    x='species',
    y='sepal_width',
    data=iris
)

plt.title('Sepal Width Distribution by Species')
plt.show()

# Custom Color Palette
custom_colors = ['#FF9F1C', '#2EC4B6', '#E71D36']

sns.set_palette(custom_colors)

sns.violinplot(
    x='species',
    y='petal_width',
    data=iris
)

plt.title('Custom Colored Petal Width Distribution')
plt.show()

# Conclusion
print("Seaborn Data Visualization Assignment Completed Successfully.")
```
